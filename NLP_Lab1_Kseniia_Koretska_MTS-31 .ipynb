## Task 1 

```python
from nltk import FreqDist
from nltk.corpus import gutenberg

text1 = nltk.Text(gutenberg.words('melville-moby_dick.txt'))
fdist = FreqDist(text1)
hapaxes = fdist.hapaxes()
less_than_5 = [word for word, freq in fdist.items() if freq < 5]

print("Hapaxes (words that occur only once):", hapaxes)
```

```python
print("Words that occur less than 5 times:", less_than_5)
```

```python
import matplotlib.pyplot as plt

text1 = nltk.Text(gutenberg.words('melville-moby_dick.txt'))
fdist = FreqDist(len(word) for word in text1)

plt.figure(figsize=(10, 6))
plt.bar(fdist.keys(), fdist.values())
plt.xlabel('Word Length')
plt.ylabel('Frequency')
plt.title('Mapping from Word Length to Word Frequency')
plt.xticks(range(max(fdist.keys()) + 1))
plt.grid(axis='y')
plt.show()
```

```python
import re

text = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
sentences = nltk.sent_tokenize(text)
word_pattern = re.compile(r'\b\w+\b')
end_words = []

for sentence in sentences:
    words = word_pattern.findall(sentence)
    if words:
        end_words.append(words[-1])

fdist = FreqDist(end_words)
most_frequent_end_words = fdist.most_common(10)

print("10 most frequent words occurring at the end of the sentence:")
for word, frequency in most_frequent_end_words:
    print(word, "-", frequency)
```

## Task 2 

```python
with open('alice_in_wonderland.txt', 'r') as f:
    raw = f.read()

tokens = nltk.word_tokenize(raw)
text = nltk.Text(tokens)

print("Concordance for the word 'Alice'")
text.concordance("Alice")
```

```python
print("\nWords similar to 'rabbit'")
text.similar("rabbit")
```

```python
print("\nDispersion plot for words 'Alice', 'rabbit', 'Queen', 'Cheshire', 'cat'")
text.dispersion_plot(["Alice", "rabbit", "Queen", "Cheshire", "cat"])
```

```python
fdist = FreqDist(text)
print("\nMost common words:")
print(fdist.most_common(10))
```

```python
def extract_monetary_amounts(text):
    pattern = r'\$[0-9]+(?:\.[0-9]{1,2})?'
    matches = re.findall(pattern, text)
    amounts = [float(match[1:]) for match in matches]
    
    return amounts

text = "first amount is $123.45, second amount is $400"
monetary_amounts = extract_monetary_amounts(text)
print("Monetary amounts found:", monetary_amounts)
```

```python
import json

json_data = '''
{
    "HTTPHost": "localhost",
    "HTTPPort": 8545,
    "Config": {
        "Enabled": true,
        "Server": "wakuv2. prod",
        "Nodes": [ "8.210.222.231", "168.166.125.145", "10.20.30.40" ]
    }
}
'''

def remove_nodes_starting_with_8(json_data):
    data = json.loads(json_data)
    data['Config']['Nodes'] = [node for node in data['Config']['Nodes'] if not re.match(r'^8\.', node)]
    updated_json_data = json.dumps(data, indent=4)
    
    return updated_json_data

updated_json_data = remove_nodes_starting_with_8(json_data)
print(updated_json_data)
```

```python
import json

json_data = '''
{
    "HTTPHost": "localhost",
    "HTTPPort": 8545,
    "Config": {
        "Enabled": true,
        "Server": "wakuv2.prod",
        "Nodes": [ "8.210.222.231", "168.166.125.145", "10.20.30.40" ]
    }
}
'''

def replace_server_entries(json_data):
    data = json.loads(json_data)
    server_name, server_type = data['Config']['Server'].split('.')
    data['Config']['Server'] = {"name": server_name, "type": server_type}
    updated_json_data = json.dumps(data, indent=4)
    
    return updated_json_data

updated_json_data = replace_server_entries(json_data)
print(updated_json_data)
```
